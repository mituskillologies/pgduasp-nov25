```python
# Create the main window
root = Tk()
root.title("Drawing Canvas")
root.geometry('1000x800')

# Create the canvas widget
canvas = Canvas(root, width=1000, height=800, bg="cyan")
canvas.pack()

# Define drawing functions
def draw_line(event):
    x1, y1 = event.x, event.y
    canvas.create_line(x1, y1, x1+2, y1+2, width=10)
    
def draw_circle(event):
    x1, y1 = event.x, event.y
    canvas.create_oval(x1 - 15, y1 - 15, x1 + 15, y1 + 15, fill="yellow")

# Bind mouse events to drawing functions
canvas.bind("<Button-3>", draw_circle)
canvas.bind("<B1-Motion>", draw_line)

# Start the main loop
root.mainloop()
```

##### Mouse operations

```python
img = cv2.imread('images/lenna.png')

def draw_circle(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x, y), 10, (255,0,255), -1)

cv2.namedWindow("image")
cv2.setMouseCallback("image", draw_circle)
while True:
    cv2.imshow("image", img)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

##### Image annotations

```python
img = cv2.imread('images/lenna.png')

text = "Maharashtra"
font = cv2.FONT_HERSHEY_COMPLEX
font_scale = 1.0
color = (200, 120, 5)
thickness = 2

# calculate the text width and height
text_size, baseline = cv2.getTextSize(text, font, font_scale, thickness)

# define text and rectangle positions
text_origin = (100, 100)
box_origin = (text_origin[0] - 10, text_origin[1] - baseline - 20)

box_width, box_height = (text_size[0] + 20, text_size[1] + 20)

cv2.rectangle(img, box_origin, (box_origin[0] + box_width,
                               box_origin[1] + box_height), 
             color, thickness)
cv2.putText(img, text, text_origin, font, font_scale, color, thickness)

cv2.imshow("Image with annotation", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
