```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
img
```

```python
threshold = 125
```

```python
img_bin = img > threshold
```

```python
img_bin
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap = 'gray')
plt.subplot(122)
plt.title("Binary Image")
plt.imshow(img_bin, cmap = 'gray')
```

##### Inverse Binary thresholding

```python
img_ibin = img < threshold
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap = 'gray')
plt.subplot(122)
plt.title("Inverse Binary Image")
plt.imshow(img_ibin, cmap = 'gray')
```

```python
ret, img_thre = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret, img_ithre = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret, img_trun = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
ret, img_zero = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(151)
plt.title("Original Image")
plt.imshow(img, cmap = 'gray')
plt.subplot(152)
plt.title("Binary Image")
plt.imshow(img_thre, cmap = 'gray')
plt.subplot(153)
plt.title("Inverse Binary Image")
plt.imshow(img_ithre, cmap = 'gray')
plt.subplot(154)
plt.title("Trunc Image")
plt.imshow(img_trun, cmap = 'gray')
plt.subplot(155)
plt.title("Tozero Image")
plt.imshow(img_zero, cmap = 'gray')
```

##### Otsu Thresholding

```python
ret, img_otsu = cv2.threshold(img, 100, 255, cv2.THRESH_OTSU)
```

```python
ret
```

```python
plt.figure(figsize=(16,9))
plt.subplot(151)
plt.title("Original Image")
plt.imshow(img, cmap = 'gray')
plt.subplot(152)
plt.title("Binary Image")
plt.imshow(img_thre, cmap = 'gray')
plt.subplot(153)
plt.title("Inverse Binary Image")
plt.imshow(img_ithre, cmap = 'gray')
plt.subplot(154)
plt.title("Trunc Image")
plt.imshow(img_trun, cmap = 'gray')
plt.subplot(155)
plt.title("Otsu Image")
plt.imshow(img_otsu, cmap = 'gray')
```
