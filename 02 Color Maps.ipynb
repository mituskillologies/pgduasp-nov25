```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread("images/lenna.png", 0)
```

##### Colormaps - Sequential

```python
plt.figure(figsize=(10,4))
plt.subplot(1,4,1)
plt.title("Gray")
plt.imshow(img_gray, cmap = "gray")
plt.subplot(1,4,2)
plt.title("Jet")
plt.imshow(img_gray, cmap = "jet")
plt.subplot(1,4,3)
plt.title("Hot")
plt.imshow(img_gray, cmap = "hot")
plt.subplot(1,4,4)
plt.title("Bone")
plt.imshow(img_gray, cmap = "bone")
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,4,1)
plt.title("Gray")
plt.imshow(img_gray, cmap = "Greys")
plt.subplot(1,4,2)
plt.title("Blue")
plt.imshow(img_gray, cmap = "Blues")
plt.subplot(1,4,3)
plt.title("Red")
plt.imshow(img_gray, cmap = "Reds")
plt.subplot(1,4,4)
plt.title("Orange")
plt.imshow(img_gray, cmap = "Oranges")
```

```python
orig_map = plt.colormaps['Blues']
rev_map = orig_map.reversed(orig_map)
```

```python
plt.imshow(img_gray, cmap = rev_map)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,4,1)
plt.title("Gray")
plt.imshow(img_gray, cmap = plt.colormaps['Greys'].reversed())
plt.subplot(1,4,2)
plt.title("Blue")
plt.imshow(img_gray, cmap = plt.colormaps['Blues'].reversed())
plt.subplot(1,4,3)
plt.title("Red")
plt.imshow(img_gray, cmap = plt.colormaps['Reds'].reversed())
plt.subplot(1,4,4)
plt.title("Orange")
plt.imshow(img_gray, cmap = plt.colormaps['Oranges'].reversed())
```

##### Diverging colormaps

```python
plt.figure(figsize=(15,8))
plt.subplot(2,4,1)
plt.title("PiYG")
plt.imshow(img_gray, cmap = plt.colormaps['PiYG'].reversed())
plt.subplot(2,4,2)
plt.title("PRGn")
plt.imshow(img_gray, cmap = plt.colormaps['PRGn'].reversed())
plt.subplot(2,4,3)
plt.title("BrBG")
plt.imshow(img_gray, cmap = plt.colormaps['BrBG'].reversed())
plt.subplot(2,4,4)
plt.title("PuOr")
plt.imshow(img_gray, cmap = plt.colormaps['PuOr'].reversed())
plt.subplot(2,4,5)
plt.title("RdBu")
plt.imshow(img_gray, cmap = plt.colormaps['RdBu'].reversed())
plt.subplot(2,4,6)
plt.title("Spectral")
plt.imshow(img_gray, cmap = plt.colormaps['Spectral'].reversed())
plt.subplot(2,4,7)
plt.title("bwr")
plt.imshow(img_gray, cmap = plt.colormaps['bwr'].reversed())
plt.subplot(2,4,8)
plt.title("seismic")
plt.imshow(img_gray, cmap = plt.colormaps['seismic'].reversed())
```

##### Cyclic color maps

```python
plt.figure(figsize=(15,8))
plt.subplot(1,4,1)
plt.title("viridis")
plt.imshow(img_gray, cmap = 'viridis')
plt.subplot(1,4,2)
plt.title("twilight")
plt.imshow(img_gray, cmap = 'twilight')
plt.subplot(1,4,3)
plt.title("hsv")
plt.imshow(img_gray, cmap = 'hsv')
plt.subplot(1,4,4)
plt.title("plasma")
plt.imshow(img_gray, cmap = 'plasma')
```
